```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('coca-cola-history.csv')
data['Date'] = pd.to_datetime(data['Date'])
sp_index = pd.read_csv('S&P Data.csv')
sp_index['Date'] = pd.to_datetime(sp_index['Date'])

#Check if data contains any missing values
# print(data.isna().any())

# print(data.info())
# print(sp_index.info())

sp_data_merged = data.merge(sp_index, on = 'Date', suffixes = ['_cola','_sp'])
sp_data_merged.head()

sp_data_merged.plot(x='Date', y = ['Close','SP500'], figsize = (12,4))
plt.xlabel('Price ($)')
plt.ylabel('Date')
plt.title('Coca-Cola VS S&P Index')
plt.show()
```

```python
data_slice = data.head(n = 250)
data_slice.plot(x = 'Date' , y = ['Open','Close', 'High','Low'] , figsize = (12,4), color = ['g','y','b','b'])
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title('Coca-Cola Line Chart')
plt.show()

plt.clf()

data_slice['SMA5'] = data_slice['Close'].rolling(5).mean()
data_slice.plot(x = 'Date', y = ['Close','SMA5'], figsize = (12,4))
plt.xlabel('Price ($)')
plt.ylabel('Date')
plt.title('Simple Moving Average 5')
plt.show()

plt.clf()

data_slice['SMA20'] = data_slice['Close'].rolling(20).mean()
data_slice.plot(x = 'Date', y = ['Close','SMA20'], figsize = (12,4))
plt.xlabel('Price ($)')
plt.ylabel('Date')
plt.title('Simple Moving Average 20')
plt.show()

plt.clf()

data_slice['SMA50'] = data_slice['Close'].rolling(50).mean()
data_slice.plot(x = 'Date', y = ['Close','SMA50'], figsize = (12,4))
plt.xlabel('Price ($)')
plt.ylabel('Date')
plt.title('Simple Moving Average 50')
plt.show()

plt.clf()

data_slice.plot(x = 'Date', y = ['Close','SMA5','SMA20', 'SMA50'], figsize = (12,4))
plt.xlabel('Price ($)')
plt.ylabel('Date')
plt.title('Simple Moving Average 5,20,50')
plt.show()

print(data_slice.head())
print(data.head())
```

```python
data['year'] = pd.DatetimeIndex(data['Date']).year
data['month'] = pd.DatetimeIndex(data['Date']).month
data['day'] = pd.DatetimeIndex(data['Date']).day
data.head()
```

```python
data_year = data.groupby('month')[['Open','Close','Low','High']].mean()
print(data_year.head())

#create figure
plt.figure(figsize = (18,8))

#define width of candlestick elements
width = 0.8
width2 = 0.1

#define up and down prices
up = data_year[data_year.Close>=data_year.Open]
down = data_year[data_year.Close<data_year.Open]

#define colors to use
col1 = 'green'
col2 = 'red'

#plot up prices
plt.bar(up.index,up.Close-up.Open,width,bottom=up.Open,color=col1) # Middle Bar
plt.bar(up.index,up.High-up.Close,width2,bottom=up.Close,color=col1) # Upper Part
plt.bar(up.index,up.Low-up.Open,width2,bottom=up.Open,color=col1) # Lower Part

#plot down prices
plt.bar(down.index,down.Close - down.Open,width,bottom=down.Open,color=col2) # Middle Bar
plt.bar(down.index,down.High - down.Open,width2,bottom=down.Open,color=col2) # Upper Part
plt.bar(down.index,down.Low - down.Close,width2,bottom=down.Close,color=col2) #Lower Part

#rotate x-axis tick labels
# plt.xticks(rotation=45, ha='right')

plt.ylabel('Price($)')
plt.xlabel('Time (Month)')
plt.title('Coca-Cola Candlestick Chart')

#display candlestick chart
plt.show()
```

```python
data_month = data.groupby('month')[['Open','Close','High','Low']].mean()
print(data_month.head())
data_month.plot(figsize = (12,4))
plt.xlabel('Date (Month)')
plt.ylabel('Price')
plt.title('Monthly Line Chart')
```
